```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf

fashion_mnist = tf.keras.datasets.fashion_mnist

(train_images, train_labels), (test_images, test_labels) = fashion_mnist.load_data()
```

```python
#Analysis of data
num_classes=10
class_names=['T-shirt','Trouser','Pullover','Dress','Coat','Heels','Shirt','Sneaker','Bag','Ankle Shoes']
```

```python
print("Train images shape: ",train_images.shape)
print("Train labels shape: ",train_labels.shape)
```

```python
#display sample images from each class
#defining the size of image
plt.figure(figsize=(10,10 ))
for i in range(num_classes):
    plt.subplot(5,5,i+1)
    plt.imshow(train_images[train_labels==i][0],cmap='gray')
    plt.title(class_names[i])
    plt.axis("off")
plt.show()
```

```python
from sklearn.model_selection import train_test_split

# Split dataset
train_images, val_images, train_labels, val_labels = train_test_split(
    train_images, train_labels, test_size=0.2, random_state=42)

print("Train images shape:", train_images.shape)
print("Validation images shape:", val_images.shape)
```

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

# Flatten images for SVM input
train_images_flat = train_images.reshape(train_images.shape[0], -1)
val_images_flat = val_images.reshape(val_images.shape[0], -1)

# Initialize SVM model
svm_model = SVC(C=1, kernel='linear', random_state=42)

# Train SVM model
svm_model.fit(train_images_flat, train_labels.ravel())

# Predict on validation set
val_predictions = svm_model.predict(val_images_flat)

# Calculate accuracy
val_accuracy = accuracy_score(val_labels, val_predictions)
print("Validation Accuracy:", val_accuracy)

from sklearn.metrics import classification_report

# Generate classification report
print(classification_report(val_labels, val_predictions, target_names=class_names))
```
